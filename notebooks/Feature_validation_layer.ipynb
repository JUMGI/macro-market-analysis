```python
from quant_research.data.systemic.loaders.systemic_loader import load_systemic_dataset
from quant_research.feature_validation.engine.validator import FeatureValidator
from quant_research.feature_validation.profiles.systemic import SystemicValidationProfile
from quant_research.feature_validation.store.store import FeatureValidationStore
from quant_research.feature_validation.configs.loader import load_validation_config
from quant_research.feature_validation.configs.validator import validate_validation_config
```

```python
bundle = load_systemic_dataset("baseline_v1")

df = bundle.data
df_z = bundle.data_z
sys_meta = bundle.metadata

config = load_validation_config("systemic_v1")
validate_validation_config(config)

profile = SystemicValidationProfile(config=config)

validator = FeatureValidator(
    profile=profile,
    dataset_id=sys_meta["name"],
    dataset_hash=sys_meta["dataset_hash"],
    config=config
)

fv_metadata = validator.validate(
    df,
    feature_info=sys_meta["features"]
)

FeatureValidationStore.save(fv_metadata)
```
