```python
# ============================================================
# IMPORTS
# ============================================================
import pandas as pd
from quant_research.config.paths import MOMENTUM_FEATURE_PATH, VOLATILITY_FEATURE_PATH
from quant_research.features.pipeline.feature_pipeline import FeaturePipeline
from quant_research.data.registry.universe_registry import get_all_assets
```

```python
# ============================================================
# CONFIGURATION
# ============================================================

ASSETS = symbols

FAMILIES = [
    "momentum",
    "volatility",
]

START_DATE = None
END_DATE = None

VALIDATE = False
OVERWRITE = True

VERBOSE = True
```

```python
# ============================================================
# PIPELINE INITIALIZATION
# ============================================================

pipeline = FeaturePipeline()
```

```python
# ============================================================
# RUN PIPELINE
# ============================================================

pipeline.run(
    assets=ASSETS,
    families=FAMILIES,
    start=START_DATE,
    end=END_DATE,
    validate=VALIDATE,
    overwrite=OVERWRITE,
    verbose=VERBOSE,
)
```
